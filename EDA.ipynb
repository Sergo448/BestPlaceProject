# Data Cleaning 

```python
#missing data

total = data.isnull().sum().sort_values(ascending=False)
percent = (data.isnull().sum() / data.isnull().count()).sort_values(ascending=False)

missing_data = pd.concat([total, percent], axis=1, keys=['total', 'percent'])
```

```python
# Можем удалить столбец Адрес, т.к. у нас есть столбец с 
# ID точки, которого нам хватает.
# И также уберем столбец Выборка, т.к. это нам сейчас не важно.
data = data.drop(columns=['Адрес', 'Выборка'])
```

```python
data = data.replace({'Парковка':{'100': "бесплатная парковка",}})
data['Парковка'].fillna('нет парквоки', inplace=True)
```

```python
# Затем у нас есть другие данные которые необходимо преобразовать.
# Тип входа, парковка, формат магазина и время работы надо преобразовать, для этого давайте закодируем их значения

from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()

data['Entranse type'] = ord_enc.fit_transform(data[['Тип входа']])
```

```python
data['Парковка'] = data['Парковка'].astype('str')
```

```python
data['Parking'] = ord_enc.fit_transform(data[['Парковка']])
```

```python
data['Magazine Format'] = ord_enc.fit_transform(data[['Формат магазина']])
```

```python
data['Working time'] = ord_enc.fit_transform(data[['Время работы']])
```

```python
data = data.drop(columns=['Тип входа', 'Парковка', 'Формат магазина', 'Время работы', ])
```

```python
data = data.replace({'Этаж':{'цоколь': 0,
                             '02.май': 2}})
```

```python
data.info()
```

```python
data['Дата закрытия'] = pd.to_numeric(pd.to_datetime(data['Дата закрытия']))
data['Дата открытия'] = pd.to_numeric(pd.to_datetime(data['Дата открытия']))
```

```python
data.info()
```

Например, в столбце "Этаж" было значение "02.май", такого допускать нельзя.
